```python
import pandas as pd
import numpy as np
```

```python
csv = pd.read_csv('fortune-1000.csv')
```

```python
df = pd.DataFrame(csv)
```

```python
df.head()
```

```python
df = df.replace('-', np.nan)
```

```python
df
```

```python
df = df.fillna(0.)
```

```python
df
```

```python
df.info()
```

```python
df['profits'] = pd.to_numeric(df['profits'])
df['employees'] = pd.to_numeric(df['employees'])
```

```python
df
```

```python
df.duplicated()
```

```python
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned = df_cleaned[['name', 'revenues', 'profits', 'assets', 'employees']]
```
